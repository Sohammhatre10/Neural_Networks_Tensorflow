# Steps in modelling the tensor

```python
import tensorflow as tf
```

```python
model = tf.keras.Sequential([
    # tf.keras.Input(shape=(1000,2)),
    tf.keras.layers.Dense(100, activation="relu"),
    tf.keras.layers.Dense(10, activation="sigmoid"),
    tf.keras.layers.Dense(1, activation='sigmoid')
])

model.compile(
    loss=tf.keras.losses.BinaryCrossentropy(),
    optimizer=tf.keras.optimizers.Adam(lr=0.1),
    metrics = ['accuracy', 'mae']
)

model.fit(X, y, epochs=100)
```

```python
model.evaluate(X,y)
```

To visualize the predictions or model boundaries we have this function

```python
import numpy as np

def plot_boundary(model, X, y):
  # Define axis boundaries of the plot and create a meshgrid
  x_min, x_max = X[:, 0].min() - 0.1, X[:,0].max() + 0.1
  y_min, y_max = X[:, 1].min() - 0.1, X[:,1].max() + 0.1
  xx,yy = np.meshgrid(np.linspace(x_min, x_max, 100),
                      np.linspace(y_min, y_max, 100))

  x_in = np.c_[xx.ravel(), yy.ravel()]

  y_pred = model.predict(x_in)

  if len(y_pred[0]) >1:
    print("Multiclass")
    y_pred = np.argmax(y_pred, axis=1).reshape(xx.shape)
  else:
    print("Binary")
    y_pred = np.round(y_pred).reshape(xx.shape)

  plt.contourf(xx,yy, y_pred, cmap='Dark2', alpha=0.7)
  plt.scatter(X[:, 0], X[:, 1], c=y, s=40, cmap='Dark2')

  plt.xlim(xx.min(), xx.max())
  plt.ylim(yy.min(), yy.max())
```

```python
plot_boundary(model, X, y)
```

```python
X_train, y_train = X[:800], y[:800]
X_test, y_test = X[800:], y[800:]
```

```python
model_2 = tf.keras.Sequential([
    tf.keras.layers.Dense(100, activation='relu'),
    tf.keras.layers.Dense(10, activation='sigmoid'),
    tf.keras.layers.Dense(1, activation='sigmoid'),
])

model_2.compile(loss='binary_crossentropy',
                optimizer=tf.keras.optimizers.Adam(lr=0.1),
                metrics=['accuracy'])

history = model_2.fit(X_train, y_train, epochs=50)

model_2.evaluate(X_test, y_test)
```

```python
model_2.predict(X_test)
```

```python
plt.figure(figsize=(12,6))
plt.subplot(1,2,1)
plot_boundary(model_2, X_train, y_train)
plt.subplot(1,2,2)
plot_boundary(model_2, X_test, y_test)
```

```python
pd.DataFrame(history.history).plot()
```

```python
tf.random.set_seed(42)

model_3 = tf.keras.Sequential([
    tf.keras.layers.Dense(100, activation='relu'),
    tf.keras.layers.Dense(10, activation='sigmoid'),
    tf.keras.layers.Dense(1, activation='sigmoid')
])

model_3.compile(
    loss='binary_crossentropy',
    optimizer='adam',
    metrics = ['accuracy']
)
lr_scheduler = tf.keras.callbacks.LearningRateScheduler(lambda epochs: 1e-4* 10**(epochs/20))

history_3 = model_3.fit(
    X_train, y_train, epochs=100, callbacks=[lr_scheduler],
)

pd.DataFrame(history_3.history).plot()
```

##### Ideal Learning Rate is near 10^-1 just where the loss decreases steeply but doesnt flatten out

```python
lrs = 1e-4 * (10**(tf.range(100)/20))
plt.figure(figsize=(10,7))
plt.semilogx(lrs, history_3.history['loss'])
plt.title("LearningRate vs Loss")
```

##### Usage of different evaluation metrics on classification models

#### `sklearn.metrics.recall_score() or tf.keras.metrics.Recall() tp/tp+fn`

Normalization

```python
train_data = train_data/train_data.max()
test_data = test_data/test_data.max()
```

```python
tf.random.set_seed(42)

model_7 = tf.keras.Sequential([
    tf.keras.layers.Flatten(input_shape=(28,28)),
    tf.keras.layers.Dense(10, activation='relu'),
    tf.keras.layers.Dense(10, activation='softmax'),
    tf.keras.layers.Dense(10, activation='softmax')
])

model_7.compile(
    loss=tf.keras.losses.SparseCategoricalCrossentropy(),
    optimizer = tf.keras.optimizers.Adam(lr=0.0016),
    metrics = ['accuracy'],
)

model_7.fit(
    train_data, train_labels, epochs=10
)
```

```python
y_pred_1 = model_7.predict(test_data)
y_pred_1 = y_pred_1.argmax(axis=1)
```

```python
conf_mat_multiclass = confusion_matrix(test_labels, y_pred_1)
```
